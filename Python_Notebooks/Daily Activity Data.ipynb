#Data Visualization

#Time Series Plot - Total Steps Over Time

```python
# Set the ActivityDate column as datetime
df['ActivityDate'] = pd.to_datetime(df['ActivityDate'])

# Group by ActivityDate and sum the TotalSteps
df_grouped = df.groupby('ActivityDate')['TotalSteps'].sum().reset_index()

# Plotly Time Series Plot
fig = px.line(df_grouped, x='ActivityDate', y='TotalSteps', title='Time Series Plot - Total Steps Over Time')

# Show the interactive Plotly plot
fig.show()
```

#Stacked Bar Chart - Total Distance with Very Active Distance

```python
df['ActivityDate'] = pd.to_datetime(df['ActivityDate'])

# Group by ActivityDate and VeryActiveDistance, summing the TotalDistance for each category
df_grouped = df.groupby(['ActivityDate', 'VeryActiveDistance'])['TotalDistance'].sum().reset_index()

# Stacked Bar Chart
fig = px.bar(df_grouped, x='ActivityDate', y='TotalDistance', color='VeryActiveDistance',
             title='Stacked Bar Chart - Total Distance with Very Active Distance',
             hover_data=['TotalDistance', 'VeryActiveDistance'])

# Show the interactive Plotly plot
fig.show()
```

#Pie Chart - Activity Distribution

```python
# Sum the activity distances
activity_distances = ['VeryActiveDistance', 'ModeratelyActiveDistance', 'LightActiveDistance', 'SedentaryActiveDistance']
df['TotalActivity'] = df[activity_distances].sum(axis=1)

# Pie Chart
pie_df = df[['VeryActiveDistance', 'ModeratelyActiveDistance', 'LightActiveDistance', 'SedentaryActiveDistance', 'TotalActivity']].sum()

fig = px.pie(pie_df, values=pie_df.values, names=pie_df.index,
             title='Pie Chart - Activity Distribution')

fig.show()
```

#Scatter Plot - Total Steps vs Total Distance

```python
# Scatter Plot
fig = px.scatter(df, x='TotalSteps', y='TotalDistance', title='Scatter Plot - Total Steps vs Total Distance',
                 labels={'TotalSteps': 'Total Steps', 'TotalDistance': 'Total Distance'},
                 hover_data=['TotalSteps','TotalDistance'])
fig.show()
```

#Box Plot - Activity Minutes Distribution

```python

# Box Plot
fig = px.box(df[['VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes']],
             title='Box Plot - Activity Minutes Distribution', labels={'variable': 'Activity Type', 'value': 'Minutes'})
fig.show()
```
