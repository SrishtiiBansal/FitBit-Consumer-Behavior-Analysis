```python
import pandas as pd
```

#Sleep Dataset

```python
df_minute_sleep=pd.read_csv("/content/drive/MyDrive/Week 8/minuteSleep_merged.csv")
df_sleep_day=pd.read_csv("/content/drive/MyDrive/Week 8/sleepDay_merged.csv")

# Merge the datasets using a common key, such as 'Id' or 'Date' depending on your data
Sleep_merged_df = pd.merge(df_minute_sleep, df_sleep_day, on='Id', how='inner')

Sleep_merged_df.head()
```

```python
# Convert to date and format
Sleep_merged_df['date'] = pd.to_datetime(Sleep_merged_df['date']).dt.strftime('%m/%d/%Y %I:%M %p')
```

```python
# Replace '/' with '-'
Sleep_merged_df['date'] = Sleep_merged_df['date'].str.replace('/', '-')
```
