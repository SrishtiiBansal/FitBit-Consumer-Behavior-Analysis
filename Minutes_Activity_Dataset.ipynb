```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#Minutes Activity dataset

```python
minute_Calories_Narrow_df=pd.read_csv("/content/drive/MyDrive/Week 8/minuteCaloriesNarrow_merged.csv")
minute_Intensities_Narrow_df=pd.read_csv("/content/drive/MyDrive/Week 8/minuteIntensitiesNarrow_merged.csv")
minute_METs_Narrow_df=pd.read_csv("/content/drive/MyDrive/Week 8/minuteMETsNarrow_merged.csv")
minute_Steps_Narrow_df=pd.read_csv("/content/drive/MyDrive/Week 8/minuteStepsNarrow_merged.csv")

# Merge datasets
merged_df = pd.merge(minute_Calories_Narrow_df, minute_Intensities_Narrow_df, on=['Id', 'ActivityMinute'], how='outer')
merged_df = pd.merge(merged_df, minute_METs_Narrow_df, on=['Id', 'ActivityMinute'], how='outer')
Minute_Activity_merged_df = pd.merge(merged_df, minute_Steps_Narrow_df, on=['Id', 'ActivityMinute'], how='outer')

Minute_Activity_merged_df.head()
```

```python
Minute_Activity_merged_df.info()
```

```python
Minute_Activity_merged_df['ActivityMinute'] = pd.to_datetime(Minute_Activity_merged_df['ActivityMinute'])
```

```python
# Convert to ActivityMinute and format
# Replace '/' with '-'

Minute_Activity_merged_df['ActivityMinute'] = pd.to_datetime(Minute_Activity_merged_df['ActivityMinute']).dt.strftime('%m/%d/%Y %I:%M %p')
Minute_Activity_merged_df['ActivityMinute'] = Minute_Activity_merged_df['ActivityMinute'].str.replace('/', '-')
```
